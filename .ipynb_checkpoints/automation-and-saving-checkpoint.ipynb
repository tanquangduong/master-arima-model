```python
# Import pmdarima as pm
import pmdarima as pm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
milk_production = pd.read_csv('./datasets/milk_production.csv', 
                    index_col='date',
                    parse_dates=True)
milk_production
```

```python
# Create auto_arima model
model1 = pm.auto_arima(milk_production,
                       seasonal=True, m=7,
                       d=0, D=1, 
                       max_p=2, max_q=2,
                       trace=True,
                       error_action='ignore',
                       suppress_warnings=True)
                       
# Print model summary
print(model1.summary())
```

```python
# Create model
model2 = pm.auto_arima(milk_production,
                      seasonal=False, 
                      d=1, 
                      trend='c',
                 	  max_p=2, max_q=2,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True) 

# Print model summary
print(model2.summary())
```

```python
# Create model for SARIMAX(p,1,q)(P,1,Q)7
model3 = pm.auto_arima(milk_production,
                      seasonal=True, m=7,
                      d=1, D=1, 
                      start_p=1, start_q=1,
                      max_p=1, max_q=1,
                      max_P=1, max_Q=1,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True) 

# Print model summary
print(model3.summary())
```

```python
# Import joblib
import joblib

# Set model name
filename = "milk_model.pkl"

# Pickle it
joblib.dump(model1, filename)
```

```python
# Load the model back in
loaded_model = joblib.load(filename)
```
